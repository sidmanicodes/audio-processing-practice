```python
# Standard imports
import numpy as np
import librosa
import os
import matplotlib.pyplot as plt
```

```python
# Load audio sample into librosa
debussy_path = os.path.join("audio", "debussy.wav")
debussy, sr = librosa.load(path=debussy_path)
```

```python
# Extract MFCCs
mfccs = librosa.feature.mfcc(y=debussy, sr=sr, n_mfcc=13)
```

```python
# Plot MFCCs
plt.figure(figsize=(25, 10))
librosa.display.specshow(
    data=mfccs,
    x_axis="time",
    sr=sr
)
plt.colorbar(format="%+2f")
plt.show()
```

```python
# Get deltas and delta-deltas
deltas = librosa.feature.delta(mfccs)
deltas2 = librosa.feature.delta(deltas, order=2)
```

```python
# Visualize deltas
plt.figure(figsize=(25, 10))
librosa.display.specshow(
    data=deltas,
    x_axis="time",
    sr=sr
)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(
    data=deltas2,
    x_axis="time",
    sr=sr
)
plt.colorbar(format="%+2f")
plt.show()
```
